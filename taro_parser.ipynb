# Прямое положение

Парсинг прямых значений карт таро

```python
result = []
url = 'https://astrometa.ru/znachenie-taro/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

articles_url =[]    

soup = BeautifulSoup(response.text, 'lxml')
for table in soup.find_all('div', class_='col-lg-2 col-md-3 col-sm-3 col-6'):
    for link in table.findAll("a", href=True):
        card = link.get('title')
        card_url = 'https://astrometa.ru/' + link.get('href')
        print(card)
        
        response = requests.get(card_url)
        soup = BeautifulSoup(response.text, 'lxml')
        pred = soup.find('p', itemprop='description').get_text() 
        pred = re.sub("([^а-яА-ЯёЁ0-9-.,])", " ", str(pred))
        
        result.append([card, pred])
        
df_straight = pd.DataFrame(result, columns=['card', 'pred'])
```

Парсинг будет проводится с сайта https://magya-online.ru

### Любовь

```python
result = []
url = 'https://magya-online.ru/znachenie_kart_taro_v_lyubvi_i_otnosheniyax'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

articles_url =[]    

soup = BeautifulSoup(response.text, 'lxml')
for table in soup.find_all('div', style='text-align: center'):
    for link in table.findAll("a", href=True):
        card = link.get('title')
        card_url = 'https://magya-online.ru' + link.get('href')
        articles_url.append(card_url)
        
        response = requests.get(card_url)
        soup = BeautifulSoup(response.text, 'lxml')
        l = []
        for i in soup.find_all('p'):
            for text in i.get_text().split('\r'):
                l.append(str(text))
        pred = ' '.join(l)
        pred = re.sub("([^а-яА-ЯёЁ0-9-.,])", " ", str(pred))    
        
        result.append([card, pred])
        
df_love = pd.DataFrame(result, columns=['card', 'pred'])
```
